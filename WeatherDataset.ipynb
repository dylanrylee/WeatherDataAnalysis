### Renaming the 'Weather' Column to a more accurate name

```python
weather_data.rename(columns= {'Weather' : 'Weather Condition'}, inplace=True)
weather_data
```

### Printing only the data when the weather was clear

```python
print(weather_data.groupby('Weather Condition').get_group('Clear'))
print(weather_data.groupby('Weather Condition').get_group('Mainly Clear'))
```

### Total number of times the weather was Clear or Mainly Clear

```python
len_clear = len(weather_data.groupby('Weather Condition').get_group('Clear'))
len_mainlyclear = len(weather_data.groupby('Weather Condition').get_group('Mainly Clear'))
print(f'Number of times that the weather was clear or mainly clear: {len_clear + len_mainlyclear}')
```

### Calculating the Expected/Mean, Variance, and Standard Deviation values of the wind speed in km/h throughout the data set

```python
weather_data.rename(columns= {'Wind Speed_km/h' : 'WindSpeed'}, inplace=True)
print("Wind Speed (km/h) Expectation, Variance, and Standard Deviation")
print(f'Expected Value: {weather_data.WindSpeed.mean()}')
print(f'Variance Value: {weather_data.WindSpeed.var()}')
print(f'Standard Deviation Value: {weather_data.WindSpeed.std()}')
```

### Calculating the Expected/Mean, Variance, and Standard Deviation values of the Relative Humidity throughout the data set

```python
weather_data.rename(columns= {'Rel Hum_%' : 'RelativeHumidity'}, inplace=True)
print("Relative Humidity Percentage Expectation, Variance, and Standard Deviation")
print(f'Expected Value: {weather_data.RelativeHumidity.mean()}')
print(f'Variance Value: {weather_data.RelativeHumidity.var()}')
print(f'Standard Deviation Value: {weather_data.RelativeHumidity.std()}')
```

### Calculating the Expected/Mean, Variance, and Standard Deviation values of the Tempearature throughout the data set

```python
weather_data.rename(columns= {'Temp_C' : 'Temperature'}, inplace=True)
print("Temperature Expectation, Variance, and Standard Deviation")
print(f'Expected Value: {weather_data.Temperature.mean()}')
print(f'Variance Value: {weather_data.Temperature.var()}')
print(f'Standard Deviation Value: {weather_data.Temperature.std()}')
```

### Printing all the instances when wind speed was above 27km/h, and Temperature was higher than the Expected Value, when the Weather Condition was rainy, and then calculating the variance of the Weather Visibility 

```python
weather_data.rename(columns= {'Visibility_km' : 'WeatherVisibility'}, inplace=True)
data = weather_data[(weather_data['WindSpeed'] > 27) & (weather_data['Temperature'] > weather_data.Temperature.mean()) & (weather_data['Weather Condition'].str.contains('Rain'))]
print(data)
print(f'\n\nVariation Value of Weather Visibility from the data above: {data.WeatherVisibility.var()}\n\n')
```
